Visualizing Vectors Using matplotlib:

We define a helper function plot_vector that uses plt.quiver to draw vectors (arrows).
We define 2D vectors v1_2d and v2_2d for visualization.
We plot the vectors and their sum to visualize vector addition.
We use plt.quiver to draw the arrows and plt.legend to add a legend.

plt.quiver: This function in matplotlib is used to plot vectors as arrows. It’s particularly useful for visualizing vector fields.

*origin: This unpacks the origin list [0, 0] into two separate arguments, 0 and 0, which represent the starting point (x, y) of the vector. In this case, the vector starts at the origin (0,0).

v[0], v[1]: These are the x and y components of the vector v. For example, if v is [3, 4], then v[0] is 3 and v[1] is 4.

angles='xy': This specifies that the angle of the arrow should be calculated with respect to the Cartesian coordinate system.

scale_units='xy': This indicates that the arrow should be scaled based on the x and y units of the plot. This ensures that the arrow length is proportional to the actual values of the vector components.

scale=1: This sets the scaling factor to 1, meaning that the length of the arrow directly corresponds to the magnitude of the vector.

**kwargs: This allows for additional keyword arguments to be passed to plt.quiver, such as color, label, etc.

plt.xlim(-10, 10): This sets the limits of the x-axis from -10 to 10. It ensures that the plot area is large enough to visualize the vectors clearly.

plt.ylim(-10, 10): This sets the limits of the y-axis from -10 to 10.

plt.grid(): This adds a grid to the plot, making it easier to visualize the position and direction of vectors relative to the coordinate system.

plt.axhline(0, color='black', linewidth=0.5): This draws a horizontal line at y=0 (the x-axis) with a thin black line. This helps to highlight the x-axis.

plt.axvline(0, color='black', linewidth=0.5): This draws a vertical line at x=0 (the y-axis) with a thin black line. This helps to highlight the y-axis.

```python
import matplotlib.pyplot as plt

def plot_vector(v, origin=[0, 0], **kwargs):
    plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, **kwargs)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.grid()
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)

# Define 2D vectors for visualization
v1_2d = np.array([3, 4])
v2_2d = np.array([1, -2])

# Plot vectors
plt.figure()
plot_vector(v1_2d, color='r', label='v1')
plot_vector(v2_2d, color='b', label='v2')
plot_vector(v1_2d + v2_2d, color='g', label='v1 + v2')

plt.legend()
plt.title("Vector Addition")
plt.show()
```

4. Gradient Descent
Example: Gradient Calculation

During the training of neural networks, gradients (which are vectors) are computed to update the weights via backpropagation. The gradient vector indicates the direction and magnitude of the steepest ascent of the loss function.

```python
import numpy as np

# Example loss function: L = (w1 - 2)^2 + (w2 + 3)^2  linear regression model coeff and interceept  1. gradient deescent 2. normal equations
def loss_function(weights):
    return (weights[0] - 2)**2 + (weights[1] + 3)**2

# Gradient of the loss function
def gradient(weights):
    return np.array([2 * (weights[0] - 2), 2 * (weights[1] + 3)])

# Initial weights
weights = np.array([0.0, 0.0])

# Learning rate
learning_rate = 0.1

# Perform one step of gradient descent
grad = gradient(weights)
weights -= learning_rate * grad


print("Updated weights:", weights)
```

```python
A = [[2, 1],
     [1, 3]]


    #  eigen value and eigen vector
# 1 0
# 0 1
     # compute eigen values and eigen vector
     # matrix can be deecomposed into eigen values and eigen vectors

# Determinant = 2*3 -1*1
# quadratic root equations

# -b+-square root(b2-4ac)/2a

# quadratic equation => a, b, c
def calculate_eigenvalues(matrix):
    # Coefficient a: Coefficient of lambda^2 in the characteristic polynomial is always 1
    a = 1

    # Coefficient b: Coefficient of lambda in the characteristic polynomial
    # It's the negative sum of the diagonal elements of the matrix
    b = -1 * (matrix[0][0] + matrix[1][1])

    # Coefficient c: Constant term in the characteristic polynomial
    # It's the determinant of the 2x2 matrix formed by non-diagonal elements of the matrix
    c = matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]

    # Solve quadratic equation to find eigenvalues
    discriminant = b ** 2 - 4 * a * c


    eigenvalue1 = (-b + discriminant ** 0.5) / (2 * a)
    eigenvalue2 = (-b - discriminant ** 0.5) / (2 * a)

    return eigenvalue1, eigenvalue2
```

```python
def calculate_eigenvectors(matrix, eigenvalues):
    eigenvectors = []
    for eigenvalue in eigenvalues:
        # Solve (A - λI)v = 0 to find eigenvector
        matrix = np.array(matrix)  # Ensure matrix is a NumPy array
        v = np.linalg.solve(matrix - eigenvalue * np.identity(len(matrix)), [0, 0])
        eigenvectors.append(v)
    return eigenvectors


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random square matrix representing a linear transformation
np.random.seed(0)
n = 2  # Change this value to change the dimension of the matrix (2 for 2x2, 3 for 3x3, etc.)
A = np.random.rand(n, n) * 10

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Form matrix Q from eigenvectors
Q = eigenvectors

# Form diagonal matrix Lambda from eigenvalues
Lambda = np.diag(eigenvalues)

# Compute the inverse of Q
Q_inv = np.linalg.inv(Q)

# Verify eigen decomposition: A = Q * Lambda * Q_inv
reconstructed_A = Q.dot(Lambda).dot(Q_inv)
assert np.allclose(A, reconstructed_A)

# Plot original matrix as arrows
plt.figure(figsize=(8, 8))
for i in range(n):
    plt.quiver(0, 0, A[0, i], A[1, i], angles='xy', scale_units='xy', scale=1, color='b', label=f'Original Matrix')

# Plot eigenvectors with lengths proportional to eigenvalues
for i in range(n):
    length = np.sqrt(eigenvalues[i]) * 5  # Adjust scaling factor for better visualization
    plt.quiver(0, 0, eigenvectors[0, i] * length, eigenvectors[1, i] * length, angles='xy', scale_units='xy', scale=1,
               color='r', label=f'Eigenvector {i+1} (Eigenvalue: {eigenvalues[i]})')

plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Matrix Transformation into Eigenvectors and Eigenvalues")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variable
x = sp.Symbol('x')

# Define the function
f_x = x**3 - 3*x**2

# Calculate the derivative
f_prime_x = sp.diff(f_x, x) # first diff
critical_points = sp.solve(f_prime_x, x) # f'(x)=0 i will get all x

# Classify critical points using the second derivative test
def classify_critical_point(x_val):
    f_double_prime = sp.diff(f_prime_x, x) # f''(x) second derivative
    second_derivative_at_x = f_double_prime.subs(x, x_val) # substitute the values in f''(x)
    if second_derivative_at_x > 0:
        return "Minimum"
    elif second_derivative_at_x < 0:
        return "Maximum"
    else:
        return "Saddle Point"

# Visualize the function and critical points
x_values = np.linspace(-1, 3, 400)
f_values = np.array([f_x.subs(x, val) for val in x_values])
plt.plot(x_values, f_values, label='f(x) = x^3 - 3x^2')
plt.scatter([val.evalf() for val in critical_points], [f_x.subs(x, val).evalf() for val in critical_points], color='red', label='Critical Points')

# Annotate critical points
for val in critical_points:
    plt.annotate(classify_critical_point(val), (val.evalf(), f_x.subs(x, val).evalf()), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function and Critical Points')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variable
x = sp.Symbol('x')

# Define the function
f_x = x**4 - 4*x**2 + 3

# Calculate the derivative
f_prime_x = sp.diff(f_x, x)

# Find the critical points (where f'(x) = 0)
critical_points = sp.solve(f_prime_x, x)

# Classify critical points using the second derivative test
def classify_critical_point(x_val):
    f_double_prime = sp.diff(f_prime_x, x)
    second_derivative_at_x = f_double_prime.subs(x, x_val)
    if second_derivative_at_x > 0:
        return "Local Minimum"
    elif second_derivative_at_x < 0:
        return "Local Maximum"
    else:
        return "Saddle Point"

# Evaluate the function at the critical points
critical_points_evaluated = [(val.evalf(), f_x.subs(x, val).evalf()) for val in critical_points]

# Evaluate the function at the endpoints of the domain
endpoints = [-2, 2]
endpoints_evaluated = [(val, f_x.subs(x, val).evalf()) for val in endpoints]

# Find the global maxima and minima
global_maximum = max(critical_points_evaluated + endpoints_evaluated, key=lambda x: x[1])
global_minimum = min(critical_points_evaluated + endpoints_evaluated, key=lambda x: x[1])

# Visualize the function and critical points
x_values = np.linspace(-2, 2, 400)
f_values = np.array([f_x.subs(x, val) for val in x_values])
plt.plot(x_values, f_values, label='f(x) = x^4 - 4x^2 + 3')

# Plot critical points
for val in critical_points_evaluated:
    plt.scatter(val[0], val[1], color='red')
    plt.annotate(f"{classify_critical_point(sp.Float(val[0]))}", val, textcoords="offset points", xytext=(0,10), ha='center')

# Plot global maxima and minima without labels
plt.scatter(global_maximum[0], global_maximum[1], color='green')
plt.scatter(global_minimum[0], global_minimum[1], color='blue')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function and Critical Points')
plt.grid(True)

# Adjust x-axis limits to ensure visibility of both critical points and global extrema
plt.xlim(-2.5, 2.5)

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
    return x**2 + y**2

# Generate input values
x_values = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced values from -5 to 5 for x
y_values = np.linspace(-5, 5, 100)  # Generate 100 evenly spaced values from -5 to 5 for y
X, Y = np.meshgrid(x_values, y_values)  # Create a grid of x and y values
Z = f(X, Y)  # Compute the values of the function f(x, y) for each (x, y) pair in the grid

# Plot the surface
fig = plt.figure()  # Create a new figure
ax = fig.add_subplot(111, projection='3d')  # Add a 3D subplot to the figure
ax.plot_surface(X, Y, Z, cmap='viridis')  # Plot the surface defined by X, Y, Z with 'viridis' colormap

# Set labels and title
ax.set_xlabel('x')  # Set label for the x-axis
ax.set_ylabel('y')  # Set label for the y-axis
ax.set_zlabel('f(x, y)')  # Set label for the z-axis
ax.set_title('Surface Plot of f(x, y) = x^2 + y^2')  # Set the title of the plot

plt.show()  # Show the plot
```
